```python
from math import sqrt
import numpy as np
from matplotlib import pyplot
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
from sklearn.datasets.samples_generator import make_regression

forcast = 10
x = make_regression(n_samples=100, n_features=1, noise=0.1)[0]
y = np.roll(x, -forcast)

x_train = x[:-(forcast + 10)]
y_train = y[:-(forcast + 10)]

x_test = x[-(forcast + 10):-forcast]
y_test = y[-(forcast + 10):-forcast]
```

```python
x_train_ = np.reshape(x_train, (x_train.shape[0], 1, x_train.shape[1]))

# design network
model = Sequential()
model.add(LSTM(4, input_shape=(1,1)))
model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')

# fit network
history = model.fit(x_train_, y_train, epochs=10, batch_size=10, verbose=2)
```

```python
# plot history
pyplot.plot(history.history['loss'], label='train')
pyplot.legend()
pyplot.show()
```

```python
x_test_ = np.reshape(x_test, (x_test.shape[0], 1, x_test.shape[1]))
predictions = model.predict(x_test_)
predictions


pyplot.plot(x)
pyplot.plot(predictions)

pyplot
```
